```python
from konlpy.tag import Mecab
from tqdm import tqdm
import pandas as pd
from glob import glob
```

```python
def content_tokenizer(df_day):
    # 한글, 숫자, 공백만 남기기
    content = content.str.replace('[^0-9a-zA-Zㄱ-ㅎㅏ-ㅣ가-힣 ]', '', regex=True)
    token_contents = []
    date_s = str(df_day['date'].values[0]).strip()
    for content in tqdm(df_day['content'], desc=date_s):
        token_content = m.morphs(content)
        token_content = ' '.join(token_content)
        token_contents.append(token_content)
    df_day['mecab'] = token_contents
    return df_day
```

```python
m = Mecab()
files = glob('../data/*.csv')
for file in files:
    df_day = pd.read_csv(file, usecols=['press', 'title', 'content', 'date', 'link'])
    df_day = content_tokenizer(df_day)
    df_day.to_csv(file, index=False)
```
